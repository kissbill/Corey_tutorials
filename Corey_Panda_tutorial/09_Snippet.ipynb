```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df.replace('NA' , np.nan, inplace=True)
df.replace('Missing' , np.nan, inplace=True)
```

```python
df.dropna()
```

```python
# axis , index row, column oszlopra dobja el az NA-t
# how , any, ha sorban bmelyik ertek hianyzik akkor az egesz sort eldobja 
# how , all ha az osszes ertek hianyzik akkor dobja el a sort csak
df.dropna(axis='index',how='all')
```

```python
# ureset ad vissza, egyik sor total NA miatt
df.dropna(axis='columns',how='any')
```

```python
df.dropna(axis='index',how='any', subset=['email'] )
```

```python
# ahol meg van a last es email azokat azok metszetet hagyja meg
df.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
# NA adatok e T/F
df.isna()
```

```python
# az osszes NA erteket helyettesiti MISSING-re
# ahol NA az ertek pl a PCA-nal akkor ott set to 0-
df.fillna('MISSING')
```

```python
# az oszlop elemeinek tipusa data type
df.dtypes
```
